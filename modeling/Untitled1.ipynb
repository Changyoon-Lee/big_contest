```python
from colorsys import hls_to_rgb 
def makecolor(n):
    c=[]
    for i in range(n+1):
        k=255/n*i
        c.append([255-k,k,0])
    return c
```

```python
import folium
def datetimeshow(date,time):
    a = df_me[df_me['date']==date]
    a = a[a['시간']==time]
    co = []
    for i in a['하차']:
        if i >3000:
            co.append(colorlist[0])
        elif i>2000:
            co.append(colorlist[1])
        elif i>1000:
            co.append(colorlist[2])
        elif i>500:
            co.append(colorlist[3])
        elif i>400:
            co.append(colorlist[4])
        elif i>300:
            co.append(colorlist[5])
        elif i>200:
            co.append(colorlist[6])
        elif i>100:
            co.append(colorlist[7])
        else:
            co.append(colorlist[8])
    a['color'] = co
    mymap = folium.Map(location=[37.558156,126.969563], tiles=None,zoom_start=12)
    for i in range(len(a)):

        lat = a['위도'].iloc[i]
        lng = a['경도'].iloc[i]
        folium.CircleMarker(location = [lat,lng], radius=10,popup=a['역명'].iloc[i],
                           color = a['color'].iloc[i], fill = True).add_to(mymap)

    folium.TileLayer('cartodbpositron').add_to(mymap)
    folium.TileLayer('openstreetmap').add_to(mymap)
    folium.TileLayer('stamentoner').add_to(mymap)
    folium.LayerControl(collapsed=False).add_to(mymap)
    return mymap
```
