```python
def prp(df):
    df['사용일자'] = pd.to_datetime(df['사용일자'], format='%Y%m%d')
    df.rename(columns={'사용일자':'date','지하철역':'역명'}, inplace=True)
    df.set_index(['역명','date'],inplace=True)
    df.columns = [5 if idx==0 else int(i[:2]) for idx, i in enumerate(df.columns)]
    return df
```
