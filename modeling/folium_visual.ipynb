```python
from colorsys import hls_to_rgb 
def makecolor(n):
    c=[]
    for i in range(n+1):
        k=255/n*i
        c.append([255-k,k,0])
    return c
```

```python
def rgb_to_hex(c):
    r, g, b = int(c[0]), int(c[1]), int(c[2])
    return '#'+hex(r)[2:].zfill(2)+hex(g)[2:].zfill(2)+hex(b)[2:].zfill(2)
```

```python
import folium
def datetimeshow(date,time):
    a = df_me[df_me['date']==date]
    a = a[a['시간']==time]
    co = []
    for i in a['하차']:
        if i >3000:
            co.append(colorlist[0])
        elif i>2000:
            co.append(colorlist[1])
        elif i>1000:
            co.append(colorlist[2])
        elif i>500:
            co.append(colorlist[3])
        elif i>400:
            co.append(colorlist[4])
        elif i>300:
            co.append(colorlist[5])
        elif i>200:
            co.append(colorlist[6])
        elif i>100:
            co.append(colorlist[7])
        else:
            co.append(colorlist[8])
    a['color'] = co
    mymap = folium.Map(location=[37.558156,126.969563], tiles=None,zoom_start=12)
    for i in range(len(a)):

        lat = a['위도'].iloc[i]
        lng = a['경도'].iloc[i]
        folium.CircleMarker(location = [lat,lng], radius=10,popup=a['역명'].iloc[i],
                           color = a['color'].iloc[i], fill = True).add_to(mymap)

    folium.TileLayer('cartodbpositron').add_to(mymap)
    folium.TileLayer('openstreetmap').add_to(mymap)
    folium.TileLayer('stamentoner').add_to(mymap)
    folium.LayerControl(collapsed=False).add_to(mymap)
    return mymap
```

```python
import folium
import folium.plugins as plugins
import numpy as np
time_index=df_me['시간'].unique().tolist()
df_hour_list=[]
for i in df_me['시간'].unique():
    df_hour_list.append(df_me.loc[df_me['시간']==i, ['위도','경도','weight']].groupby(['위도','경도','weight']).sum().reset_index().values.tolist())
    
mymap = folium.Map(location=[37.558156,126.969563], tiles='cartodbpositron',zoom_start=12)
plugins.HeatMapWithTime(df_hour_list, radius=12, index=time_index, auto_play=True,
    gradient={0.1: 'blue', 0.4: 'lime', 0.6: 'orange', 0.9: 'red'}, min_opacity=0.5, max_opacity=1, use_local_extrema=True).add_to(mymap)
mymap
```
