```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

m = 1
s = 10
m2 = 1
s2 = 6

x = np.linspace(m-3*s, m+3*s, 100)
y = stats.norm.pdf(x, m, s)
x2 = np.linspace(m-3*s, m+3*s, 100)
y2 = stats.norm.pdf(x2, m2, s2)
plt.plot(x, y)
plt.plot(x2, y2)
plt.show()
```
