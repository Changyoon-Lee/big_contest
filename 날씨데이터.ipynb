```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
pages = [1, 2, 3, 4, 5]
url = 'http://www.weather.go.kr/weather/climate/past_cal.jsp?stn=108&yy=2020&mm={}&obs=1'.format(pages[0])

resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'lxml')
tag = soup.select('tr td.align_left')
tag
```

```python
tag_li = []
for i in range(len(tag)):
    tag_li.append(tag[i].text)
```

```python
while '\xa0' in tag_li:    
    tag_li.remove('\xa0') # '\xa0' 삭제
tag_li
```

# 날씨데이터 (다운로드)

```python
weather = pd.read_csv('날씨데이터(기상자료개방포털).csv', encoding='cp949')
```

```python
weather = weather[['일시', '기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)']]
weather.columns = ['date', '기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)']
```

```python
import datetime
weather['date'] = pd.to_datetime(weather['date'])
```

```python
weather.info()
```

```python
# weather['시간'] = ''
for i in range(len(weather)):
#     weather['시간'][i] = weather['date'][i].hour
    weather['date'][i] = weather['date'][i].date()
```

```python
# 전체 데이터프레임도 시간 앞 숫자 따서 변경할 것!
weather
```
